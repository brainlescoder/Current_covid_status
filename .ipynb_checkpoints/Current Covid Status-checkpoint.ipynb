### INDIAN GOV DATA OF COVID

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

```python
df.shape
```

```python
df=df.iloc[:-1,1:]
```

```python
df.tail(15)
```

```python
df['active']
```

```python
max(df['positive'])
```

### Total Positive Cases In India 

```python
plt.figure(figsize = (12, 9))

plt.barh(df['state_name'], df['positive'], color = 'r', alpha = 0.5)
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
```

```python
plt.figure(figsize=(12, 9))

plt.barh(df['state_name'], df['new_active'], color = 'g', alpha = 0.5)
     
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
```

```python
plt.figure(figsize=(12, 9))
plt.barh(df['state_name'], df['death'], color = 'b', alpha = 0.5)
    
plt.xlabel('Deaths')  
plt.title('Corona Virus Cases')  
```
